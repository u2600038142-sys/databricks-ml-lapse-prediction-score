```python
CATALOG     = "lapse_scoring_dev"
SRC_SCHEMA  = "02_silver"
SRC_TABLE   = "fwd_max_master_category_bn_silver"
DST_SCHEMA  = "03_gold"
DST_TABLE   = "fwd_max_master_category_gold"

SRC_FQN = f"{CATALOG}.`{SRC_SCHEMA}`.{SRC_TABLE}"
DST_FQN = f"{CATALOG}.`{DST_SCHEMA}`.{DST_TABLE}"

# pastikan schema Gold ada
spark.sql(f"CREATE SCHEMA IF NOT EXISTS {CATALOG}.`{DST_SCHEMA}`")

df = spark.table(SRC_FQN)

(df.write
   .mode("overwrite")              # replace isi tabel tujuan
   .option("mergeSchema","true")   # jaga-jaga jika ada evolusi skema
   .option("overwriteSchema","true")
   .format("delta")
   .saveAsTable(DST_FQN))

print(f"Copied OK -> {DST_FQN} (rows={spark.table(DST_FQN).count()})")
display(spark.sql(f"SELECT * FROM {DST_FQN} LIMIT 20"))
```
